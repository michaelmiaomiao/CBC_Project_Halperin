```r
set.seed(123)
df=new_data[sample(1:nrow(new_data),50000),]

index2  <- createDataPartition(df$count, p = 0.8, list = FALSE)
use <- df[index2, ]
valid  <- df[-index2, ]
```

```r
#df=new_data
set.seed(123)
index3  <- createDataPartition(use$count, p = 0.75, list = FALSE)
train <- use[index3, ]
test  <- use[-index3, ]
```

```r
dim(train)

dim(test)

dim(valid)
```

```r
set.seed(123)
model1 <- train(
  count ~., data = train, method = "glmnet",
  trControl = trainControl("cv", number = 10),
  tuneLength= 10
)
```

```r
model1$bestTune
```

```r
coef_table=coef(model1$finalModel, model1$bestTune$lambda)
```
