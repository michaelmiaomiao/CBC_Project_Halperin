```r
set.seed(123)
index   <- sample(1:nrow(data), round(nrow(data) * 0.7))
train <- data[index, ]
test  <- data[-index, ]
```

```r
linear_regression = lm(formula = Ozone ~ ., data = train)
```

```r
linear_regression %>% summary()
```

```r
lm_pred=predict(linear_regression,test,interval = "prediction")
lm_pred
```

```r
mse <- function(sm) 
    mean(sm$residuals^2)
mse(linear_regression)
```

```r
tt=as.vector(lm_pred[,1])
```

```r

pred.lm.int <- sapply(1:nrow(lm_pred), function(i) {
  tmp <- lm_pred[i] + rnorm(1001, 0, sqrt(mse(linear_regression)))
  quantile(tmp, c(0.025, 0.975))
})
t(pred.lm.int)
```

```r

set.seed(123)
tree_bag <- train(Ozone ~ ., data = train, 
                 method = 'treebag',
                 trControl = fitControl,
                 
                 verbose = FALSE)
tree_bag
tree_bag_pred= predict(tree_bag,test,predict.all=TRUE)
tree_bag_pred[1:3]
pred.tree.int <- sapply(1:35, function(i) {
  tmp <- tree_bag_pred[i] + rnorm(1001, 0, min(tree_bag$results$RMSE))
  quantile(tmp, c(0.025, 0.975))
})
t(pred.tree.int)

```

```r

set.seed(123)
bay_net <- train(Ozone ~ ., data = train, 
                 method = 'brnn',
                 trControl = fitControl,
                 
                 verbose = FALSE)

bay_net_pred= predict(bay_net,test,predict.all=TRUE)

```
